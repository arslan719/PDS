```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv('/data_clean/cleaned_data.csv')
df
```

```python
df['Frailty'] = df['Frailty'].astype(str)
df_y = df[df['Frailty'] == 'Y']
df_y
```

```python
df_n = df[df['Frailty'] == 'N']
df_n
```

```python
df_max = df.groupby(['Frailty'], sort=False).max()
df_max
```

```python
df_min = df.groupby(['Frailty'], sort=False).min()
df_min
```

```python
df_max_age = df.groupby(['Frailty'], sort=False)['Age'].max().reset_index()
df_max_age
```

```python
df_min_age = df.groupby(['Frailty'], sort=False)['Age'].min().reset_index()
df_min_age
```

```python
df_max_weight = df.groupby(['Frailty'], sort=False)['Weight '].max().reset_index()
df_max_weight
```

```python
df_min_weight = df.groupby(['Frailty'], sort=False)['Weight '].min().reset_index()
df_min_weight
```
