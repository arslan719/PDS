```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/data_clen/cleaned_data.csv")
```

```python
#sns.pairplot(data=df, hue='gender')
#plt.show()

avg_math_scores = df.groupby('gender')['math score'].mean().reset_index()

# Create a bar plot to compare average math scores by gender
plt.figure(figsize=(6, 4))
custom = {'male': 'skyblue', 'female': 'pink'}
sns.barplot(data=avg_math_scores, x='gender', y='math score', palette=custom)
plt.xlabel('Gender')
plt.ylabel('Average Math Score')
plt.title('Average Math Scores by Gender')
plt.show()
```

This analysis shows the Average maths score by gender given in the data



```python
sns.countplot(data=df, x='parental level of education', palette='Set3')
plt.xticks(rotation=45)
plt.show()
```

This analysis shows the parental education information for acquired degree


```python
plt.hist(df['writing score'], bins=20, edgecolor='k', color= 'red')
plt.xlabel('Writing Score')
plt.ylabel('Frequency')
plt.title('Distribution of Writing Scores')
plt.show()
```

This is the distribution of writing scores along x-axis are writing scores and on y-axis is the count of how many student obtained those marks


```python
df_new = pd.DataFrame(df)

# Calculate the average reading score for each race/ethnicity group
avg_scores = df_new.groupby('race/ethnicity')['reading score'].mean().reset_index()

# Create a bar plot to display average reading scores
plt.figure(figsize=(8, 6))
sns.barplot(data=avg_scores, x='race/ethnicity', y='reading score', palette='Set1')
plt.xticks(rotation=45)
plt.xlabel('Race/Ethnicity')
plt.ylabel('Average Reading Score')
plt.title('Average Reading Scores by Race/Ethnicity')
plt.show()
```

This is the reading scores by race/ethnicity by groups distribution

```python
# Create a scatter plot to compare math scores and writing scores
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='math score', y='writing score', hue='gender', palette='Set2')
plt.xlabel('Math Score')
plt.ylabel('Writing Score')
plt.title('Comparison of Math and Writing Scores by Gender')
plt.legend(title='Gender')
plt.show()
```

This is comparison of maths score by gender
