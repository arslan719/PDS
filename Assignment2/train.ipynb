```python
import pandas as pd
import numpy as np
import datetime
```

```python
df = pd.read_csv('train.csv')
```

```python
# Initialize empty lists to store statistics
columns = []
unique_values = []
missing_percentage = []
data_types = []

# Loop through the columns
for col in df.columns:
    columns.append(col)
    unique_values.append(df[col].nunique())
    missing_percentage.append(df[col].isna().sum()/ len(df))
    data_types.append(df[col].dtype)

# Create a summary DataFrame
summary_df = pd.DataFrame({
    'Column': columns,
    'Unique Values': unique_values,
    'Missing Percentage': missing_percentage,
    'Data Type': data_types
})

print(summary_df)
```

# Dropping the columns

Dropping the column "Unnamed" since it looks like a counter and not providing the exact count of rows. It is counting the rows as 6018 when the real count is 5847 and it is not useful for analysis.

Dropped the column "New_Price" since there is not enough data in it. It has about 86% null values though it won't be very helpful by imputing (replacing with mean, mode). So at this point dropping it is a good option.


```python
drop = ["Unnamed: 0" , "New_Price"]
df = df.drop(columns= drop)
```

```python
df.count()
```

```python
df['Engine'] = pd.to_numeric(df['Engine'].str.replace(r'[^\d.]', '', regex=True), errors='coerce')
df['Power'] = pd.to_numeric(df['Power'].str.replace(r'[^\d.]', '', regex=True), errors='coerce')
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
df.count()
```

```python
def convert_to_float(value):
    if isinstance(value, str):
        value = value.replace(r'[^\d.]', '', regex=True)
        return float(value)
    return float(value)

# Apply the custom function to the 'Mileage' column
df['Mileage'] = df['Mileage'].apply(convert_to_float)
df['Engine'] = df['Engine'].apply(convert_to_float)
df['Power'] = df['Power'].apply(convert_to_float)
```

```python
df
```

For one hot encoding pandas dummie function will create indicators for all catagories

```python
df = pd.get_dummies(df, columns=['Fuel_Type', 'Transmission'])
```

```python
df
```

```python
# Calculate the current year
current_year = datetime.datetime.now().year

# Create a new column "Car_Age" by subtracting "Year" from the current year
df['Car_Age'] = current_year - df['Year']
```
